```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

df = pd.read_csv(r'C:\Users\Negligent\Desktop\ABTS\movies.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df=df.dropna()
df
```

```python
df.dtypes
```

```python
df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
df
```

```python
newDF = df['released'].str.split(" ",n=3,expand=True)
newDF.rename(columns={0:"Month",1:"Day",2:"Year",3:"Country"},inplace=True)
df['correctedYear']=newDF['Year']
df
```

```python
df.sort_values(by=['gross'],inplace = False, ascending = False)
```

```python
pd.set_option('display.max_rows', 50)
```

```python
#Scatter plot: Budget v Gross

plt.scatter(x=df['budget'],
           y=df['gross'])
plt.title('Budget v Gross')
plt.xlabel('Budget')
plt.ylabel('Gross')
plt.show()
```

```python
df.corr()
```

```python
import seaborn as sns
```

```python
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix,annot=True)
plt.title('Visual Correlations')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```
